```python
import pygame as pg
import numpy as np

pg.init()

# Tamanho da tela e definição do FPS
screen = pg.display.set_mode((640, 480))
clock = pg.time.Clock()
FPS = 60  # Frames per Second
 
BLACK = (0, 0, 0)
COR_PERSONAGEM = (200, 200, 200)

# Inicializar posicoes
s0 = np.array([80,190])
a = np.array([0, 0.05])
y = pg.mouse.get_pos()
v = y - s0
v = v/np.linalg.norm(v)
v *= 5
s = s0

# Personagem
personagem = pg.Surface((5, 5))  # Tamanho do personagem
personagem.fill(COR_PERSONAGEM)  # Cor do personagem

canhao = pg.Surface((5, 5))  # Tamanho do personagem
canhao.fill((100,100,100))  # Cor do personagem

corpo_celeste = pg.Surface((20, 20))  
corpo_celeste.fill((30,50,100))  # Cor do personagem

pg.display.set_caption('image')
luci = pg.image.load("images/luci.png").convert()
luci = pg.transform.scale(luci,(50,50))
luci_retangulo = pg.Surface((50, 50))  # Tamanho do personagem
luci_retangulo.fill((200,200,200))  # Cor do personagem

pg.display.set_caption('image')
imp = pg.image.load("images/anjo.png").convert()
imp = pg.transform.scale(imp,(70,48))

rodando = True
pressed = True

x_satelite = np.random.randint(150,490)
y_satelite = np.random.randint(150,330)

y_luci = np.random.randint(150,330)

while rodando:
    # Capturar eventos
    for event in pg.event.get():
        if event.type == pg.QUIT:
            rodando = False
        if event.type == pg.MOUSEBUTTONDOWN:
            pressed = True

    if pressed:
        if s[0]<10 or s[0]>630 or s[1]<10 or s[1]>470: # Se eu chegar ao limite da tela, reinicio a posição do personagem
            y = pg.mouse.get_pos()
            v = y - s0
            v = v/np.linalg.norm(v)
            v *= 5
            s = s0

    # Controlar frame rate
    clock.tick(FPS)

    # Processar posicoes
    v = v 
    s = s + v

    # Desenhar fundo
    screen.fill(BLACK)

    # Desenhar personagem
    rect = pg.Rect(s, (10, 10))  # First tuple is position, second is size.
    screen.blit(personagem, rect)

    # rect_canhao = pg.Rect(s, (10, 10))  # First tuple is position, second is size.
    # screen.blit(canhao, rect_canhao)

    celeste = pg.draw.circle(screen, (0, 200, 200), (x_satelite,y_satelite), 20)  # First tuple is position, second is size.
    # screen.blit(corpo_celeste, celeste)
    
    # Using blit to copy content from one surface to other
    screen.blit(imp, (10, 180))


    rect_luci = pg.Rect((570, y_luci), (50,50))  # First tuple is position, second is size.
    screen.blit(luci_retangulo, rect_luci)

    screen.blit(luci, (570, y_luci))

    # Update!
    pg.display.update()
    pressed = False

# Terminar tela
pg.quit()
```
